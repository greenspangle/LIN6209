## Generate and display the plot

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python
plt.plot(ypoints, '*:g')
plt.show()
```

```python
plt.plot(ypoints, 's--b', ms= 10, mec='m', mfc='g')
plt.show()
```
